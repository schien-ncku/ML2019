```python
import pandas as pd
import numpy as np

df = pd.read_csv("train_updated.csv", encoding="utf-8", usecols=range(3, 27))
train_data = df.replace(['NR'], [0.0])
```

```python
# convert csv data to array
array = np.array(train_data).astype(float)
```

```python
x_list = []
y_list = []

for i in range(0, 4320, 18):
    for j in range(24-9):
        mat = array[i:i+18, j:j+9]
        label = array[i+9, j+9]
        x_list.append(mat)
        y_list.append(label)
```

```python
x = np.array(x_list)
y = np.array(y_list)
```

```python
# start trying linear regression
bias = 0
weight = 1
rate = 0.01

epoch = 20
for e in range(epoch):
    loss = 0
    diff = 0
    w_diff = 0
    for i in range(3200):
        for j in range(9):
            y_pred = bias + weight*x_train[i, 9, j]
            delta = y_pred - y_train[i] 
            diff += delta
            w_diff += delta*x_train[i,9,j]
            loss += delta**2

    # update bias: sum of difference
    b_gradient = diff/(3200*9)
    new_bias = bias - rate*b_gradient
    #new_bias = 0
    # update weight: sum of vector length
    w_gradient = w_diff/(3200*9)
    new_weight = weight - rate*rate*w_gradient
    #new_weight = 1

    average_loss = loss/(3200*9)
    #average_loss = np.sqrt(loss/(3200*9))
    
    print('epoch ', e, '; bias: ', bias, '; weight: ', weight, '; loss: ', average_loss)
    
    bias = new_bias
    weight = new_weight
```
