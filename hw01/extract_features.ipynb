```python
array = np.array([['AMB_TEMP'], 
                 ['CH4'], 
                 ['CO'], 
                 ['NHMC'], 
                 ['NO'], 
                 ['NO2'], 
                 ['NOx'], 
                 ['O3'], 
                 ['PM10'], 
                 ['PM2.5'], 
                 ['RAINFALL'], 
                 ['RH'], 
                 ['SO2'], 
                 ['THC'], 
                 ['WD_HR'], 
                 ['WIND_DIREC'], 
                 ['WIND_SPEED'], 
                 ['WS_HR']])

with open('train_updated.csv') as f:
    count = 0
    for line in f:
        index = count%18
        if 1==index :
            temp = line.split(',')
            l = len(temp)
            for x in range(3, l) :
                array[index-1][x-3]=temp[x]
            
        count=count+1

print(array[0][1])
```
