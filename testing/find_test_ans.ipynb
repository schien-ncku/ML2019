```python
import pandas as pd
import numpy as np

testdata = pd.read_csv("given/test_X.csv")
pm2_5_test = testdata[testdata['AMB_TEMP']=='PM2.5'].iloc[:,2:]
```

```python
x_test = np.array(pm2_5_test)
x_test
```

```python
alldata = pd.read_csv("original_data.csv")
```

```python
alldata = alldata.replace(['NR'], [0.0])
alldata = alldata.replace([''],[0.0])
alldata
```

```python
pm2_5_all = np.array(alldata[alldata['測項']=='PM2.5'].iloc[:,3:])
```

```python
pm2_5_all
```

```python
pm2_5_all[0]
```

```python
d = np.zeros((240,3))

for i in range (240):
    a = x_test[i,:]
    f = 0
    for j in range (365):
        for k in range (24-9):
            b = pm2_5_all[j,k:k+9]
            if np.array_equal(a,b) :
                print(i, ": ", j, "-", k)
                d[i,0] = j
                d[i,1] = k
                f = 1
                break
        if (f == 1):
            break
```

```python
for i in range (240):
    if d[i,0]>0:
        ii = int(d[i,0]*18+9)
        l = int(d[i,1])
        d[i,2] = alldata.iloc[ii,l+10] # this is the answer
        print(i, ") PM2.5 on ", alldata.iloc[ii,0], " at hour ", l+10, " is ", int(d[i,2]))
```
